Load API key and create .env

```python
from getpass import getpass
from dotenv import load_dotenv
import os

key = getpass("Paste your OPENAI_API_KEY (input hidden): ")
with open(".env", "w") as f:
    f.write(f"OPENAI_API_KEY={key}\n")
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
print("API Key loaded:", OPENAI_API_KEY is not None)
```

Initialize OpenAI client

```python
from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)
```

Define embedding function

```python
def get_embeddings(text_list):
    response = client.embeddings.create(
        model="text-embedding-3-small",
        input=text_list
    )
    return [e.embedding for e in response.data]
```

Define report generation function

```python
def generate_sleep_report(sleep_efficiency, caffeine, alcohol, exercise):
    prompt = f"""
    You are a sleep health expert.
    Based on this data:
    - Sleep Efficiency: {sleep_efficiency}%
    - Caffeine Intake: {caffeine} cups/day
    - Alcohol Intake: {alcohol} drinks/week
    - Exercise Frequency: {exercise} hours/week

    Write a short, friendly recommendation for improving sleep.
    """
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=200
    )
    return response.choices[0].message.content
```

Test


```python
sample_text = ["I slept 6 hours with 3 cups of coffee"]
sample_embedding = get_embeddings(sample_text)
print("Sample embedding length:", len(sample_embedding[0]))

sample_report = generate_sleep_report(75, 3, 2, 1)
print("\nSample Sleep Report:\n", sample_report)
```
